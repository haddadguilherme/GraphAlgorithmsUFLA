## Match Máximo Bipartido

Um *match* em um gráfico bipartido é um conjunto de arestas escolhidas de forma que duas arestas não compartilhem um ponto final. Um *match* máximo (MM) é um *match* do tamanho máximo (número máximo de arestas). Em um MM, se alguma aresta for adicionada a ela, ela não será mais uma *match*. Pode haver mais de um MM para um determinado gráfico bipartido.

Existem muitos problemas do mundo real que podem ser formados como *match* Bipartido. Como esse da atividade 3 que busca uma forma de fazer um *match* de aplicações a serem executadas e computadores disponíveis.

# O Algoritmo


Vamos primeiro definir os formulários de entrada e saída. A entrada é na forma de matriz de Edmonds, que é uma matriz 2D '*bpGraph [M] [N]*' com M linhas (para M aplicações candidatas a utilizar o computador) e N colunas (para N computadores) e nas duas últimas posições são passados os números de pretenções de uso e o nome da aplicação. O valor *bpGraph [i] [j]* é 1 se o requerente está interessado em *j'ésimo* computador, caso contrário, 0. Output é a lista da sequência correta de utilização dos computadores ou *!* caso não seja possível encontrar o *match*.

Em *bpm()*, nós, um a um, tentamos todos os computadores em que um candidato 'u' está interessado até encontrarmos um *match*, ou todos os computadores são tentados sem sorte. Para cada computador que tentamos, seguimos.

Se um computador não é atribuído a ninguém, simplesmente o atribuímos a aplicação e retornamos verdadeiro. Se um computador é atribuído a outra aplicação, digamos x, então, recursivamente, verificamos se x pode receber algum outro computador. Para garantir que x não receba o mesmo computador novamente, marcamos o computador 'v' como visto antes de fazermos chamadas recursivas para x. Se x puder obter outro computador, nós mudamos a aplicação para o computador 'v' e retornamos verdadeiro. Usamos uma matriz maxR [0..N-1] que armazena as aplicações atribuídas a diferentes computadores.

Se *bmp()* retornar true, significa que há um caminho de aumento na rede de fluxo e uma unidade de fluxo é incluída no resultado em *maxBPM()*.

Caso o número de correspondências seja igual ao número solicitado pela matriz de entrada, marcamos o *status = Verdadeiro*, caso contrário marcamos como *Falso*

Dessa forma, ao fim é possível mostrar o resultado da forma solicitada.
