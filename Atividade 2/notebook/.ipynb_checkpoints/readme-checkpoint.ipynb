### Floyd-Warshall

Na teoria de grafos, o algoritmo de Floyd-Warshall é um algoritmo que resolve o problema de calcular o caminho mais curto entre todos os pares de vértices em um grafo valorado (com peso). Dado uma matriz de adjacência, o algoritmo percorre essa matriz em busca de vértices intermediários que minimizem o caminho de um vértice A para um vértice B - (A -> X -> B), sendo "X" um ou mais vértices intermediários. Quando esta situação ocorre, a matriz de adjacência é atualizada com a menor distância recém descoberta.

## Problema Proposto e Ideia de Resolução

Adamantina é uma cidade em que qualquer ponto da cidade pode ser alcançável a partir de outro por meio de um automóvel
(em ambos os sentidos). Cada quarteirão conecta duas interseções (esquinas). As ruas são de mão dupla. Dadas duas
interseções A e B em Adamantina, a distância entre A e B é dada pelo número mínimo de quarteirões que devem
ser percorridos de A para B. A Figura 1 ilustra esta ideia a partir de uma parte da cidade (as interseções
estão marcadas por um círculo vermelho e os blocos por quadrados pretos).
Recentemente, devido a trechos carregados de intenso tráfego de veículos, os engenheiros da prefeitura
sugeriram uma reformulação no layout de sua rede de transportes. A ideia é transformar algumas ruas de mão
dupla em ruas de mão única. Esta proposta deve ser planejada de maneira cautelosa, pois o prefeito atual
não quer perder em hipótese alguma a acessibilidade. Por outro lado, mesmo que o alcance seja mantido, é
possível que as distâncias entre interseções específicas possam ser significativamente aumentadas. A Figura
2 apresenta um exemplo de alteração sugerida.
Após diversas assembléias, a prefeitura decidiu aceitar as propostas que aumentem as distâncias entre quaisquer interseções em um fator x, acrescido de uma constante y. Por exemplo: se as interseções u e v possuem distância d(u,v) , então com a nova proposta, a distância não poderá
passar de xd(u,v) + y.





### Solução Desenvolvida:

1. Primeiramente são criadas 2 matrizes quadradas de ordem N, sendo N o número de interseções. Todos valores dessas matrizes são definidos como valores inteiros "infinitos".

2. Os valores das distâncias entre as interseções (esquinas) no **projeto atual** de Adamantina são calculados e inseridos em uma das matrizes com valores infinitos criada anteriormente, substituindo os valores anteriores nas posições que representam as interseções do projeto atual.

3. Os valores das distâncias entre as interseções (esquinas) no **projeto sugerido** de Adamantina são calculados e inseridos na outra matriz com valores infinitos criada anteriormente, substituindo os valores anteriores nas posições que representam as interseções no projeto sugerido.

4. Posteriormente são calculadas as menores distâncias possíveis (Floyd-Warshall) entre todos os pares de esquinas de ambos os projetos e atualizados esses valores nas matrizes de seus respectivos projetos.

5. Por fim, é comparado a diferença das distâncias entre as mesmas posições nas duas matrizes, utilizando o fator de tolerância fornecido pela prefeitura. Se todas diferenças estiverem dentro do limite de tolerância, o projeto sugerido é aceito, caso contrário ele é rejeitado.
