### Floyd-Warshall

Na teoria de grafos, o algoritmo de Floyd-Warshall  um algoritmo que resolve o problema de calcular o caminho mais curto entre todos os pares de vrtices em um grafo valorado (com peso). Dado uma matriz de adjacncia, o algoritmo percorre essa matriz em busca de vrtices intermedirios que minimizem o caminho de um vrtice A para um vrtice B - (A -> X -> B), sendo "X" um ou mais vrtices intermedirios. Quando esta situao ocorre, a matriz de adjacncia  atualizada com a menor distncia recm descoberta.

## Problema Proposto e Ideia de Resoluo

Adamantina  uma cidade em que qualquer ponto da cidade pode ser alcanvel a partir de outro por meio de um automvel
(em ambos os sentidos). Cada quarteiro conecta duas intersees (esquinas). As ruas so de mo dupla. Dadas duas
intersees A e B em Adamantina, a distncia entre A e B  dada pelo nmero mnimo de quarteires que devem
ser percorridos de A para B. A Figura 1 ilustra esta ideia a partir de uma parte da cidade (as intersees
esto marcadas por um crculo vermelho e os blocos por quadrados pretos).
Recentemente, devido a trechos carregados de intenso trfego de veculos, os engenheiros da prefeitura
sugeriram uma reformulao no layout de sua rede de transportes. A ideia  transformar algumas ruas de mo
dupla em ruas de mo nica. Esta proposta deve ser planejada de maneira cautelosa, pois o prefeito atual
no quer perder em hiptese alguma a acessibilidade. Por outro lado, mesmo que o alcance seja mantido, 
possvel que as distncias entre intersees especficas possam ser significativamente aumentadas. A Figura
2 apresenta um exemplo de alterao sugerida.
Aps diversas assemblias, a prefeitura decidiu aceitar as propostas que aumentem as distncias entre quaisquer intersees em um fator x, acrescido de uma constante y. Por exemplo: se as intersees u e v possuem distncia d(u,v) , ento com a nova proposta, a distncia no poder
passar de xd(u,v) + y.





### Soluo Desenvolvida:

1. Primeiramente so criadas 2 matrizes quadradas de ordem N, sendo N o nmero de intersees. Todos valores dessas matrizes so definidos como valores inteiros "infinitos".

2. Os valores das distncias entre as intersees (esquinas) no **projeto atual** de Adamantina so calculados e inseridos em uma das matrizes com valores infinitos criada anteriormente, substituindo os valores anteriores nas posies que representam as intersees do projeto atual.

3. Os valores das distncias entre as intersees (esquinas) no **projeto sugerido** de Adamantina so calculados e inseridos na outra matriz com valores infinitos criada anteriormente, substituindo os valores anteriores nas posies que representam as intersees no projeto sugerido.

4. Posteriormente so calculadas as menores distncias possveis (Floyd-Warshall) entre todos os pares de esquinas de ambos os projetos e atualizados esses valores nas matrizes de seus respectivos projetos.

5. Por fim,  comparado a diferena das distncias entre as mesmas posies nas duas matrizes, utilizando o fator de tolerncia fornecido pela prefeitura. Se todas diferenas estiverem dentro do limite de tolerncia, o projeto sugerido  aceito, caso contrrio ele  rejeitado.
